```

doc_file2=nlp(open('arnav.txt').read())
```

```
#Function to convert text into sentences and tokenize each sentence into list
def sent_to_words(orgfile):
    varfile=list(orgfile.sents)
    for sentence in varfile:    
        yield(gensim.utils.simple_preprocess(str(sentence),deacc=True,min_len=3))
        
data_input=list(sent_to_words(doc_file2))
data_input
```

```
#create dictionary: To create dicitonary need to create list and not string
dict=corpora.Dictionary(data_input)
print(dict.token2id)
```

```
model=gensim.models.Word2Vec(data_input,size=300,sg=3,min_count=1,workers=3)
```

```
model.train(data_input, total_examples=len(data_input), epochs=200)
```
